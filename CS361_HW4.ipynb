```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of trials
trials = 10**5
p_values = np.linspace(0.1, 1, 10)

seats = 10
def tickets_needed(p):
    return np.ceil(seats / p)


tickets_required = [tickets_needed(p) for p in p_values]
ticket_data = pd.DataFrame({"p (probability)": p_values, "Tickets Needed": tickets_required})
ticket_data

```

```python
def simulate_flight(p):
    passengers = np.random.binomial(seats, p, trials)  # Number of passengers who turn up
    women = np.random.binomial(passengers, 0.5, trials)  # Assuming half of the passengers are women
    flights_with_two_women = women >= 2
    passengers_if_flight = passengers[flights_with_two_women]  # Passengers only for flights that go
    return np.mean(passengers_if_flight)

# Perform the simulation for different values of p
expected_passengers_given_flight = [simulate_flight(p) for p in p_values]

# Plot the results
plt.figure(figsize=(10,6))
plt.plot(p_values, expected_passengers_given_flight)
plt.title("Expected Number of Passengers Given Flight Operates (At Least Two Women)")
plt.xlabel("Probability p of Passenger Turning Up")
plt.ylabel("Expected Number of Passengers")
plt.grid(True)
plt.show()
```
