```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python


file = pd.read_csv('https://raw.githubusercontent.com/limu050404/CS361-Homework/main/nuclear-plants.txt', sep='\t', skipinitialspace=True) # Use tab as the delimiter and skip initial whitespace
file
```

```python
from scipy import stats
```

```python
z_scores_cost = stats.zscore(file['Cost'])
z_scores_cost
```

According to the output, when considering Cost, all the costs are not outliers because they all have a z-score whose absolute value is smaller than 3. Let's see what will happen when it turns to megawatts.

```python
z_scores_megawatt = stats.zscore(file['Mwatts'])
z_scores_megawatt
```

According to the output, even the first nuclear power plant whose megawatt output seems to be an outlier, all those datas have a z-score whose absolute value is smaller than 3, so there's no outliers.

```python
mean_cost = file['Cost'].mean()
mean_cost
```

```python
std_cost = file['Cost'].std()
std_cost
```

The mean cost of a power plant is approximately 461.56, and its standard deviation is approximately 170.12.

```python
cost_per_megawatt = file['Cost']/file['Mwatts']
mean_cost_per_megawatt = cost_per_megawatt.mean()
mean_cost_per_megawatt
```

```python
std_cost_per_megawatt = cost_per_megawatt.std()
std_cost_per_megawatt
```

The mean cost per megawatt is approximately 0.57, and the standard deviation is 0.187

```python
plt.hist(cost_per_megawatt)
plt.xlabel('Cost Per Megawatt')

plt.show()
```

```python
median_cost_per_megawatt = cost_per_megawatt.median()
median_cost_per_megawatt
```

The mean value is smaller than the median value, so although it is hard to judge from the graph, it is skewed to the left.
