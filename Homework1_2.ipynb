```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('https://raw.githubusercontent.com/limu050404/CS361-Homework/main/hot-dogs-1.txt', sep='\t')
file
```

```python
plt.hist(file[file['Type'] == 'Beef']['Sodium'], alpha=0.5, label='Beef')
plt.hist(file[file['Type'] == 'Meat']['Sodium'], alpha=0.5, label='Meat')
plt.hist(file[file['Type'] == 'Poultry']['Sodium'], alpha=0.5, label='Poultry')
plt.xlabel('Sodium')
plt.ylabel('Frequency')
plt.title('Sodium Content by Hotdog Type')
plt.legend()
plt.show()

plt.hist(file[file['Type'] == 'Beef']['Calories'], alpha=0.5, label='Beef')
plt.hist(file[file['Type'] == 'Meat']['Calories'], alpha=0.5, label='Meat')
plt.hist(file[file['Type'] == 'Poultry']['Calories'], alpha=0.5, label='Poultry')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.title('Calorie Content by Hotdog Type')
plt.legend()
plt.show()
```
