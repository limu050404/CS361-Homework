```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('https://raw.githubusercontent.com/limu050404/CS361-Homework/main/Readability.csv')
file
```

```python
plt.boxplot([file[file['GROUP'] == 1]['3SYL'], file[file['GROUP'] == 2]['3SYL'], file[file['GROUP'] == 3]['3SYL']], labels=['Group 1', 'Group 2', 'Group 3'])
plt.xlabel('Group')
plt.ylabel('3SYL')
plt.title('Boxplot of 3 or more syllable words by Group')
plt.show()
```

Group 3 has some outliers of syllable word used in AD, while group 1 amd 2 don't have. Group 3's range of 3 or more syllable word is significantly lower than the other two groups, which may indicate that the readers had a comparatively lower level of education because those words are simplier. Group one have the widest range (after removing all the outliers).

```python
plt.boxplot([file[file['GROUP'] == 1]['SEN'], file[file['GROUP'] == 2]['SEN'], file[file['GROUP'] == 3]['SEN']], labels=['Group 1', 'Group 2', 'Group 3'])
plt.xlabel('Group')
plt.ylabel('SEN')
plt.title('Boxplot of number of sentences of an advertisement by Group')
plt.show()
```

The average number of sentences of adv of three groups doesn't differ so much. Group 3 still has some outliers while others don't. Group 2 has the lowest mean, but its maximum number of sentences is the largest among three groups. The narrowest range and the low value of sentences of group 3 could prove that group 3 is the lowest educational level.
